## Padding

```python
import numpy as np
from keras.utils import to_categorical
from keras.preprocessing.sequence import pad_sequences # menambah padding, agar sama panjangnya
```

```python
import json

# Load token2idx and idx2token
with open('token2idx.json', 'r') as f:
    token2idx = json.load(f)

with open('idx2token.json', 'r') as f:
    idx2token = json.load(f)

# Load tag2idx and idx2tag
with open('tag2idx.json', 'r') as f:
    tag2idx = json.load(f)

with open('idx2tag.json', 'r') as f:
    idx2tag = json.load(f)
```

```python
# menambah padding agar panjang token sama
def padd(data):
  tokens = data['token_encode'].tolist()
  maxlen = 40 # maximal panjang token untuk dijadikan patokan padding
  pad_tokens = pad_sequences(tokens, maxlen=maxlen, dtype='int32', padding='post') # menambahkan padding agar sama panjang

  n_tags = len(tag2idx)
  tags = data['label_encode'].tolist()
  pad_tags = pad_sequences(tags, maxlen=maxlen, dtype='int32', padding='post', value= tag2idx["O"])
  pad_tags = [to_categorical(i, num_classes=n_tags) for i in pad_tags]

  return pad_tokens, pad_tags
```

```python
# data_train padding
pad_tokens, pad_tags = padd(data_train, df)
pad_tags = np.array(pad_tags)
```

## Build Model

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, TimeDistributed
from keras.layers import Bidirectional
from keras import Model, Input
import tensorflow as tf
from tensorflow import keras
```

```python
input_dims = len(list(set(df['token'].to_list()))) + 1 # jumlah fitur/ variabel independen unik yang akan diproses model
output_dims = len(list(set(df['label'].to_list()))) # jumlah variabel depend yang akan dihasilkan
input_lenghts = 40 # tentukan urutan paling panjang dari dataset => 34
n_tags = len(tag2idx)
print(input_dims, output_dims, input_lenghts, n_tags)
```

```python
# model1

input_word = Input(shape = (input_lenghts,))
model = Embedding(input_dim=input_dims, output_dim=output_dims, input_length=input_lenghts)(input_word)
model = Bidirectional(LSTM(units=output_dims, return_sequences=True))(model)
output = TimeDistributed(Dense(n_tags,  activation='softmax'))(model)
model = Model(input_word, output)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
import re
# Ensure UNK token is in token2idx dictionary
if '<UNK>' not in token2idx:
    token2idx['<UNK>'] = len(token2idx)
    idx2token[len(idx2token)] = '<UNK>'

# Example tokenizer that uses a predefined dictionary of multi-word phrases
def custom_tokenizer(text, phrases):
    phrases = sorted(phrases, key=len, reverse=True)
    phrase_patterns = [re.escape(phrase) for phrase in phrases]
    pattern = re.compile('|'.join(phrase_patterns))
    tokens = pattern.findall(text)
    return tokens

# Example usage
phrases = dictionary['Data'].to_list()
#text = "Akhir Tahun Astra Agro Lestari AALI Targetkan Produksi TBS Naik"
tokens = custom_tokenizer(text, phrases)

# Mapping and padding
tokens_idx = [[token2idx.get(w, token2idx['<UNK>']) for w in tokens]]
prediksi_kata_padded = pad_sequences(sequences=tokens_idx, maxlen=maxlen, padding="post", value=n_token - 1)

# Predict
preds = np.argmax(model.predict(np.array(prediksi_kata_padded)), axis=-1)[0]

# Convert predictions to tag names
predicted_tags = [idx2tag[p] for p in preds[:len(tokens)]]  # ensure length matches tokens

# Prepare readable output
result = list(zip(tokens, predicted_tags))
print("Tokenized Input and Predicted Tags:")
for token, tag in result:
    print(f"{token}: {tag}")
```
